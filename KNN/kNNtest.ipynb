```
import numpy as np
import pylab as pl
import scipy
from scipy.spatial.distance import cdist
%pylab inline
np.random.seed(166)
data = np.random.randn(100,2)
labels = np.random.randint(0, 2, 100)
labels
```

```
testparticle = np.random.randn(2)
testparticle
```

```
ax = pl.figure(figsize=(10,10)).add_subplot(111)
ax.axis('equal')

ax.plot(data[:,0][labels==0], data[:,1][labels==0], 'o')
ax.plot(data[:,0][labels==1], data[:,1][labels==1], 'x')
ax.plot(testparticle[0], testparticle[1], 'k+', ms=30, lw=3)
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4);
ax.set_xlabel("x")
ax.set_ylabel("y");
```

```
cdist(np.atleast_2d(testparticle), data)
```

```
d = cdist(np.atleast_2d(testparticle), data)
d
```

```
k = 1
labeltest = labels[np.argmin(d)]
nn = data[np.argmin(d)]
```

```
ax = pl.figure(figsize=(10,10)).add_subplot(111)
ax.axis('equal')

ax.plot(data[:,0][labels==0], data[:,1][labels==0], 'o')
ax.plot(data[:,0][labels==1], data[:,1][labels==1], 'x')
c = 'IndianRed'
if labeltest == 0:
  c = 'SteelBlue'
ax.plot(testparticle[0], testparticle[1], 'k+', ms=30, lw=3, color=c)
ax.plot(nn[0], nn[1], 'k*', lw=3)
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4);
```

```
k = 5
nns = data[np.argsort(d)][0][:k]
label = labels[np.argsort(d)][0][:k]
labeltest = 0 if (label == 0).sum() > (label == 1).sum() else 1
```

```
ax = pl.figure(figsize=(10,10)).add_subplot(111)
ax.axis('equal')

ax.plot(data[:,0][labels==0], data[:,1][labels==0], 'o')
ax.plot(data[:,0][labels==1], data[:,1][labels==1], 'x')
c = 'IndianRed'
if labeltest == 0:
  c = 'SteelBlue'
ax.plot(testparticle[0], testparticle[1], 'k+', ms=30, lw=10, color=c)
ax.plot(nns[:,0][labels[np.argsort(d)][0][:k]==0], 
        nns[:,1][labels[np.argsort(d)][0][:k]==0], 'ko', lw=3, alpha = 0.5)
ax.plot(nns[:,0][labels[np.argsort(d)][0][:k]==1], 
        nns[:,1][labels[np.argsort(d)][0][:k]==1], 'k+', lw=3, alpha = 0.5)
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4);
```
